```python
import pandas as pd
import numpy as np 
```

```python
data_set_ready = pd.read_csv("dataset_scaled_no_outliers.csv", sep=",")
data_set_ready
```

```python
data_set_ready.groupby("RainTomorrow").size()
class_feature = data_set_ready["RainTomorrow"].values
previous_feature = data_set_ready.drop(columns="RainTomorrow").values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

x_treino, x_teste, y_treino, y_teste = train_test_split(previous_feature, class_feature, test_size = 0.3, random_state = 0)
# Separando os dados em folds
kfold = KFold(n_splits = 30, shuffle=True, random_state = 5)
```

```python
from sklearn.naive_bayes import GaussianNB
naive = GaussianNB()
naive.fit(x_treino, y_treino)
previsoes_naive = naive.predict(x_teste)
previsoes_naive
print(accuracy_score(y_teste, previsoes_naive))
print(confusion_matrix(y_teste, previsoes_naive))
print(classification_report(y_teste, previsoes_naive))

modelo = GaussianNB()
resultado = cross_val_score(modelo,previous_feature, class_feature, cv = kfold)
print("Acurácia Média: %.2f%%" % (resultado.mean() * 100.0))
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='poly', random_state=1, C = 2)
svm.fit(x_treino, y_treino)
previsoes_svm = svm.predict(x_teste)
previsoes_svm
print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_svm) * 100.0))
print(confusion_matrix(y_teste, previsoes_svm))
print(classification_report(y_teste, previsoes_svm))


#modelo = SVC(kernel='rbf', random_state=1, C = 2)
#resultado = cross_val_score(modelo, previous_feature, class_feature, cv = kfold)

# Usamos a média e o desvio padrão
#print("Acurácia Média: %.2f%%" % (resultado.mean() * 100.0))
```

```python
from sklearn.linear_model import LogisticRegression
logistica = LogisticRegression(random_state=42, max_iter=200, penalty="l2",
                               tol=0.0001, C=1,solver="lbfgs")
logistica.fit(x_treino, y_treino)
previsoes_logistica = logistica.predict(x_teste)

print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_logistica) * 100.0))
print(confusion_matrix(y_teste, previsoes_logistica))
print(classification_report(y_teste, previsoes_logistica))

modelo = LogisticRegression(random_state=42, max_iter=200, penalty="l2",
                               tol=0.0001, C=1,solver="lbfgs")
resultado = cross_val_score(modelo,previous_feature, class_feature, cv = kfold)
print("Acurácia Média: %.2f%%" % (resultado.mean() * 100.0))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7, metric='minkowski', p=1, weights="distance")
knn.fit(x_treino, y_treino)
previsoes_knn = knn.predict(x_teste)

print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_knn) * 100.0))
print(confusion_matrix(y_teste, previsoes_knn))
print(classification_report(y_teste, previsoes_knn))

resultado = cross_val_score(knn,previous_feature, class_feature, cv = kfold)
print("Acurácia Média: %.2f%%" % (resultado.mean() * 100.0))
```

```python
from sklearn.tree import DecisionTreeClassifier

arvore = DecisionTreeClassifier(criterion='gini', random_state = 43, max_depth=6)
arvore.fit(x_treino, y_treino)
previsoes_arvore = arvore.predict(x_teste)

print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_arvore) * 100.0))
print(confusion_matrix(y_teste, previsoes_arvore))
print(classification_report(y_teste, previsoes_arvore))


resultado = cross_val_score(arvore,previous_feature, class_feature, cv = kfold)
print("Acurácia Média: %.2f%%" % (resultado.mean() * 100.0))
```

```python
from sklearn.ensemble import RandomForestClassifier

random = RandomForestClassifier(n_estimators=50, criterion='gini', random_state = 42, max_depth=6)
random.fit(x_treino, y_treino)
previsoes_random = random.predict(x_teste)

print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_random) * 100.0))
print(confusion_matrix(y_teste, previsoes_random))
print(classification_report(y_teste, previsoes_random))

resultado = cross_val_score(random,previous_feature, class_feature, cv = kfold)
print("Acurácia Média: %.2f%%" % (resultado.mean() * 100.0))
```

```python
from xgboost import XGBClassifier
xg = XGBClassifier(max_depth=6, learning_rate=0.005, n_estimators=300, objective='binary:logistic', random_state=3)
xg.fit(x_treino,y_treino)
previsoes_xg = xg.predict(x_teste)

print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_xg) * 100.0))
print(confusion_matrix(y_teste, previsoes_xg))
print(classification_report(y_teste, previsoes_xg))

#resultado = cross_val_score(xg,previous_feature, class_feature, cv = kfold)
#print("Acurácia Média: %.2f%%" % (resultado.mean() * 100.0))
```
